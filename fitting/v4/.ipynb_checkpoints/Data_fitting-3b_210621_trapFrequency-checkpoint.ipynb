From master thesis: $\omega^2 = \frac{\text{Re}(\alpha) 2P_0}{m\epsilon_0 S w_0^2 c} = \frac{\text{Re}(\alpha) 4P_0}{m\epsilon_0\pi w_0^4 c}$ since $S=\frac{\pi}{2}w_0^2$ for a gaussian beam

```python
def trapFrequency_x(ODT1, ODT2, ODT3):
    beamwaist_12 = 42*10**(-6)
    beamwaist_3 = 35*10**(-6)
    trapFrequency_0DT1 = 180*polarizability_0*4*ODT_power.ODT_voltage_to_power(1, ODT1)/(m*epsilon_0*np.pi*beamwaist_12**4*c)
    trapFrequency_0DT2 = 180*polarizability_0*4*ODT_power.ODT_voltage_to_power(2, ODT2)/(m*epsilon_0*np.pi*beamwaist_12**4*c)
    trapFrequency_0DT3 = 180*polarizability_0*4*ODT_power.ODT_voltage_to_power(3, ODT3)/(m*epsilon_0*np.pi*beamwaist_3**4*c)
    print(np.sqrt(trapFrequency_0DT1))
    print(np.sqrt(trapFrequency_0DT2))
    print(np.sqrt(trapFrequency_0DT3))
    return np.sqrt(trapFrequency_0DT1 + trapFrequency_0DT2 + trapFrequency_0DT3)
```
