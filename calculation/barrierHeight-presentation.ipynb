# 2D-2b formulation

We start with a 2-body Schrödinger equation in 3D. The two atoms of equal mass are trapped in a harmonic confinement of frequency $\omega_z$ $\left(V_{ho}(z_1) = \frac{1}{2}m\omega_z^{2}z_1^2\right)$ and interact via a long range interactionn potential $V_{dd}$ and a short range Van der Walls one. We look for stationnary solutions:
\begin{equation}
\hat H \Psi = E\Psi \\
\hat H = \left(\frac{\hat{\vec{p}}_1^2}{2m} + \frac{\hat{\vec{p}}_2^2}{2m} + V_{ho}(z_1) + V_{ho}(z_2) + V_{dd}(\rho_1, \rho_2, z_1, z_2)\right)
\end{equation}

## Relative coordinate system

Going to a relative coordinate system $\vec{R} = \frac{\vec{r}_1 + \vec{r}_2}{2}$ ; $\vec{r} = \vec{r}_2 - \vec{r}_1$ ; $\hat{\vec{P}} = \frac{\hat{\vec{p}}_1 + \hat{\vec{p}}_2}{2} = M\dot{\vec{R}}$ ; $\hat{\vec{p}} = \hat{\vec{p}}_2 - \hat{\vec{p}}_1 = \mu\dot{\vec{r}}$ ; with $M=2m$ and $\mu = m/2$, it gives:

\begin{equation}
\frac{\hat{\vec{p}}_1^2}{2m} + \frac{\hat{\vec{p}}_2^2}{2m} = \frac{1}{2m}\left(\frac{\hat{\vec{P}}^{2}+4\hat{\vec{p}}^{2}}{2}\right) = \frac{\hat{\vec{P}}^2}{2M} + \frac{\hat{\vec{p}}^2}{2\mu}
\end{equation}
\begin{equation}
V_{ho}(z_1) + V_{ho}(z_2) = \frac{1}{2}m\omega_z^2 z_1^2 + \frac{1}{2}m\omega_z^2 z_2^2 = \frac{1}{2}m\omega_z^2 \left(\frac{4Z^2+z^2}{2}\right) = \frac{1}{2}M\omega_z^2 Z^2 + \frac{1}{2}\mu\omega_z^2 z^2
\end{equation}
\begin{equation}
V_{dd}(\rho_1, \rho_2, z_1, z_2) = \frac{\mu_0}{4\pi}(10\mu_B)^2\frac{1-3\cos^2\theta}{r^3} = \frac{\mu_0}{4\pi}(10\mu_B)^2\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}
\end{equation}

I will only keep the relative part of the Hamiltonian and write:
\begin{equation}
\hat H_{\text{rel}} \Psi = E_{\text{rel}}\Psi \\
\end{equation}
With:
\begin{equation}
\hat H_{\text{rel}} = \frac{\hat{\vec{p}}^2}{2\mu} + \frac{1}{2}\mu\omega_z^2 z^2 + \frac{\mu_0}{4\pi}(10\mu_B)^2\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}
\end{equation}
I will omit the $_{\text{rel}}$ subscript in the following.

## Reducing to length scale

I will reuse Ticknor's definition for the following length scale:
\begin{equation}
\boxed{a_{dd} = \frac{\mu_0}{4\pi}(10\mu_B)^2\frac{\mu}{\hbar^2} = 10.3\text{ nm}}
\end{equation}
The value is calculated for $^{162}$Dy which has a magnetic moment of $9.93\mu_B$ and a mass of 161.92 amu.
\begin{equation}
\boxed{a_z = \sqrt{\frac{\hbar}{\mu\omega_z}}}
\end{equation}
The oscillator length is equal to 30 nm at $\omega_z = 140$ kHz.
The dipolar energy:
\begin{equation}
\boxed{E_{dd} = \frac{\hbar^2}{\mu a_{dd}^2} = k_B\cdot 56\text{ }\mu\text{K}}
\end{equation}

The Hamiltonian can now be written as:
\begin{equation}
\hat H = \frac{\hbar^2}{\mu}\left(-\frac{\nabla^2}{2} + \frac{z^2}{2a_z^4} + a_{dd}\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}\right) = \frac{\hbar^2}{\mu}\hat{h}
\end{equation}
And now \[\hat{h}\Psi = \frac{\mu E}{\hbar^2}\Psi\]

## Expanding the wavefunction

The wavefunction is $2\pi$ periodic for the variable $\theta$. So we can expand it like:
\begin{equation}
\Psi(\rho, \theta, z) = \sum_m e^{im\theta}\psi_m(\rho, z)
\end{equation}
It is further useful to write $\psi_m(\rho, z) = \tilde{\psi}_m(\rho, z)/\sqrt{\rho}$ and then further expand the wavefunction on the harmonic oscillator basis $\left{\chi_n\right}$:
\begin{equation}
\Psi(\rho, \theta, z) = \sum_m e^{im\theta}\sum_n\chi_n(z)\frac{\phi_{m,n}(\rho)}{\sqrt{\rho}}
\end{equation}

In cylindrical coordinates, the Laplacian reads:
\begin{equation}
\nabla^{2}\Psi = \frac{1}{\rho}\frac{\partial}{\partial \rho}\left(\rho\frac{\partial\Psi}{\partial\rho}\right) + \frac{1}{\rho}\frac{\partial}{\partial \theta}\left(\frac{1}{\rho}\frac{\partial\Psi}{\partial\theta}\right) + \frac{\partial}{\partial z}\left(\frac{\partial\Psi}{\partial z}\right)
\end{equation}
For the $\rho$ derivative:
\begin{equation}
\frac{1}{\rho}\frac{\partial}{\partial \rho}\left(\rho\frac{\partial}{\partial\rho}\left(\frac{\phi_{m,n}(\rho)}{\sqrt{\rho}}\right)\right) = \frac{1}{\rho}\frac{d}{d\rho}\left[\rho\left( \frac{\phi^{'}_{m,n}(\rho)}{\sqrt{\rho}}-\frac{1}{2}\rho^{-3/2}\phi_{m,n}(\rho) \right) \right] = \frac{1}{\sqrt{\rho}}\left(\phi_{m,n}^{''}(\rho)+\frac{1}{4\rho^2}\phi_{m,n}(\rho)\right)
\end{equation}

Therefore:
\begin{equation}
\hat{h}\Psi = -\frac12\left\{\underbrace{\frac{1}{\sqrt{\rho}}\sum_m e^{im\theta}\left[\sum_n\chi_n(z)\left(\phi^{''}_{m,n}(\rho)+\frac{1}{4\rho^{2}}\phi_{m,n}(\rho)\right)\right]}_{\rho \text{ derivative} } - \underbrace{\frac{m^{2}}{\rho^{2}}\Psi}_{\theta \text{ derivative} } + \underbrace{\sum_m e^{im\theta}\sum_n\chi^{''}_n(z)\frac{\phi_{m,n}(\rho)}{\sqrt{\rho}}}_{z \text{ derivative}} \right\} + \frac{z^2}{2a_z^4}\Psi + a_{dd}\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}\Psi
\end{equation}

### Harmonic oscillator part

By definition of the harmonic oscillator wavefunction:
\begin{equation}
-\frac12\chi^{''}_n (z)+\frac{z^2}{2a_z^4}\chi_n(z) = \left(n+\frac12\right)\frac{\hbar\omega_z}{\hbar^2/\mu}\chi_n(z)
\end{equation}
Which gives:
\begin{equation}
\hat{h}\Psi = \frac{1}{\sqrt{\rho}}\sum_m e^{im\theta}\left[\sum_n\chi_n(z)\left(-\frac12\frac{d^2}{d\rho^2}\cdot + \frac{m^2 - 1/4}{2\rho^2}+\left(n+\frac12\right)\frac{1}{a_z^2} + a_{dd}\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}\right)\phi_{m,n}(\rho)\right] = \frac{\mu E}{\hbar^2}\Psi
\end{equation}

Now I can multiply by $\int e^{-im^{'}\theta}\chi_{n^{'}}^*(z)dz$ to give:
\begin{equation}
\frac{1}{\sqrt{\rho}}\left\{\left(-\frac12\frac{d^2}{d\rho^2}\cdot + \frac{m^{'2} - 1/4}{2\rho^2}+\left(n^{'}+\frac12\right)\frac{1}{a_z^2}\right)\phi_{m^{'},n^{'}}(\rho) + \sum_n\phi_{m^{'},n}(\rho)a_{dd}\underbrace{\int\chi^*_{n^{'}}(z)\frac{1-3\frac{z^2}{z^2+\rho^2}}{(z^2+\rho^2)^{3/2}}\chi_n(z)dz}_{U_{n^{'},n}(\rho)}\right\} = \frac{\mu E}{\hbar^2}\frac{\phi_{m^{'},n^{'}}(\rho)}{\sqrt{\rho}}
\end{equation}

Switching back to $m^{'}\rightarrow m$ and $n^{'} \rightarrow n$
\begin{equation}
\boxed{\left(-\frac12\frac{d^2}{d\rho^2}\cdot + \frac{m^2 - 1/4}{2\rho^2}+\left(n+\frac12\right)\frac{1}{a_z^2}\right)\phi_{m,n}(\rho) + \sum_{n^{'}}U_{n, n^{'}}(\rho)\phi_{m,n^{'}}(\rho) = \frac{\mu E}{\hbar^2}\phi_{m,n}(\rho)}
\end{equation}

### Dimensioneless equation

We rescale the distances with $a_{dd}$ and the energies with the dipolar energy $E_{dd} = \frac{\hbar^2}{\mu a_{dd}^2}$ (note that the definition differs from Ticknor's paper as I think he made a mistake). $\tilde{\rho} = \rho/a_{dd} ; \tilde{z} = z/a_{dd} ; D = a_{dd}/a_z$. There is a subtlety for rescaling $U_{n, n^{‘}}$ as the harmonic oscillator wavefunction $\chi_n(z) = \frac{1}{\sqrt{2^n n!}}\left(\frac{1}{a_z^2\pi}\right)^{1/4}e^{-\frac{z^2}{2a_z^2}}H_n(z) = \sqrt{1/a_{dd}}\frac{1}{\sqrt{2^n n!}}\sqrt{D}\left(\frac{1}{\pi}\right)^{1/4}e^{-\tilde{z}^2 D^2/2}H_n(\tilde{z}D) = \sqrt{1/a_{dd}}\tilde{\chi}_{n, D}(\tilde{z})$. Which gives that:
\begin{equation}
U_{n, n^{‘}}(\rho) = \frac{D}{a_{dd}}\frac{1}{a_{dd}^2}a_{dd}\int\tilde{\chi}^*_{n}(\tilde{z})\frac{1-3\frac{\tilde{z}^2}{\tilde{z}^2+\tilde{\rho}^2}}{(\tilde{z}^2+\tilde{\rho}^2)^{3/2}}\chi_{n^{'}}(\tilde{z})d\tilde{z} = \frac{1}{a_{dd}^2}\tilde{U}_{n, n^{‘}}(\tilde{\rho})
\end{equation}
And the full equation reads:
\begin{equation}
\boxed{\left(-\frac12\frac{d^2}{d\tilde{\rho}^2}\cdot + \frac{m^2 - 1/4}{2\tilde{\rho}^2}+\left(n+\frac12\right)D^2\right)\phi_{m,n}(\rho) + \sum_{n^{'}}\tilde{U}_{n, n^{'}}(\tilde{\rho})\phi_{m,n^{'}}(\rho) = \frac{E}{E_D}\phi_{m,n}(\rho)}
\end{equation}

### Vectorial formulation

It can all be written like a vectorial equation, with $\tilde{U}(\tilde{\rho})$ being a matrix $\forall m$:
\begin{equation}
\boxed{\left\{\left(-\frac12\frac{d^2}{d\tilde{\rho}^2}\cdot + \frac{m^2 - 1/4}{2\tilde{\rho}^2}-\left(\frac{E}{E_D}-D^2\frac12\right)\right)\mathbb{1} + D^2 diag(n) + \tilde{U}(\tilde{\rho})\right\}\vec{\phi}_{m}(\rho) = 0}
\end{equation}
Where the part in brackets is what we want to diagonalize to find uncoupled and non-zero solutions of the equation. 

## Numerical solution

For the following I will take atoms approching from far away with zero extra kinetic energy, just the bare harmonic oscillator zero-point energy, so $E/E_D = D^2/2$ and the equation is:
\begin{equation}
\left\{\left(-\frac12\frac{d^2}{d\tilde{\rho}^2}\cdot + \frac{m^2 - 1/4}{2\tilde{\rho}^2}\right)\mathbb{1} + D^2 diag(n) + \tilde{U}(\tilde{\rho})\right\}\phi_{m,n}(\rho) = 0
\end{equation}

### Side remark on adding the C6

The potential $V$ is measured in units of $\frac{\hbar^{2}}{2\mu (1\text{nm})^{2}}$. The oscillator length is $a_{z} = \sqrt{\frac{\hbar}{\mu\omega_{z}}}$, the dipole length is $a_{dd} = \frac{\mu_{0}}{4\pi}(10\mu_{B})^{2}\frac{m}{\hbar^{2}} = \frac{\mu_{0}}{2\pi}(10\mu_{B})^{2}\frac{\mu}{\hbar^{2}} = 20.86 \text{ nm}$ and the Van der Waals length is, as in Julienne's paper (https://journals.aps.org/rmp/pdf/10.1103/RevModPhys.82.1225) $r_{vdw} = \frac{1}{2}\left(2\mu C_{6}/\hbar^{2}\right)^{1/4} = 4.26\text{ nm}$ if I take Nascimbene's paper (https://iopscience.iop.org/article/10.1088/1361-6455/50/1/014005/pdf) where $C_{6} = 2273 E_{h}a_{0}^{6}$.

This transforms the equation:
\[
V(\rho, z) = \frac{1}{2}\mu\omega_{z}^{2}z^{2} + \frac{\hbar^{2}(m^{2}-1/4)}{2\mu\rho^{2}}-\frac{C_{6}}{r^{6}}+\frac{\mu_{0}}{4\pi}(10\mu_{B})^{2}\frac{1-3\cos^{2}\theta}{r^{3}}
\]
into
\[
V(\rho, z) = \frac{V}{\frac{\hbar^{2}}{2\mu (1\text{nm})^{2}}} = \frac{1}{a_{z}^{4}}z^{2} + \frac{m^{2}}{\rho^{2}}-\frac{(2r_{vdw})^{4}}{(z^{2}+\rho^{2})^{3}}+a_{dd}\frac{1-3\frac{z^{2}}{z^{2}+\rho^{2}}}{(z^{2}+\rho^{2})^{3/2}}
\]

```python
from scipy.optimize import root
from scipy.integrate import quad
from scipy.signal import argrelextrema

import numpy as np
import matplotlib.pyplot as plt
```

```python
def matrixCentrifugal(rho, m, N):
    return np.identity(N)*(m**2-1/4)/(2*rho**2)

def matrixEnergy(E, N):
    return np.identity(N)*E

def hermite(z, n):
    hermiteFactors = np.zeros(n+1)
    hermiteFactors[n] = 1
    hermite_n = np.polynomial.hermite.Hermite(hermiteFactors)
    return hermite_n(z)

def harmonicOscillatorWF(z, D, n):
    return 1/np.sqrt(2**n*np.math.factorial(n))*np.sqrt(D)*np.pi**(-1/4)*np.exp(-0.5*(z*D)**2)*hermite(z*D, n)

def harmonicOscillatorProba(z, D, n):
    return harmonicOscillatorWF(z, D, n)**2

def dipoleInteraction(z, rho):
    return (1-3*z**2/(z**2+rho**2))/(z**2+rho**2)**(3/2)

def interactionMatrixElement(rho, D, n, nprim):
    def integrant(z, rho, n, nprim):
        return harmonicOscillatorWF(z, D, n)*harmonicOscillatorWF(z, D, nprim)*dipoleInteraction(z, rho)
    return quad(integrant, -np.infty, np.infty, args = (rho, n, nprim))

def interactionMatrix(rho, D, N):
    matrix = np.zeros((N, N))
    for n in range(N):
        for nprim in range(N):
            matrix[n, nprim] = interactionMatrixElement(rho, D, n, nprim)[0]
    return matrix

def harmonicMatrix(N):
    diag = np.array(range(N))
    return np.diag(diag)

def nonDiagonalMatrix(rho, D, N):
    return interactionMatrix(rho, D, N) + D**2*harmonicMatrix(N)

def diagonalisation(rho, D, N):
    matrix = nonDiagonalMatrix(rho, D, N)
    return np.linalg.eigh(matrix)

def effectivePotential(rho, D, m, N):
    v, P = diagonalisation(rho, D, N)
    return (m**2-1/4)/(2*rho**2) + v[0]  # maybe I should multiply everything by 2 to match the d2/drho2
```

## Integrating the potential along z

```python
def V(rho, z, m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy):
    return (1/a_z)**4*z**2 + (m**2-1/4)/rho**2-(2*r_vdw)**4/(z**2+rho**2)**3+a_dd*(1-3*(z**2/(z**2+rho**2)))/(z**2+rho**2)**(3/2)

def Psi0(z, a_z):
    return (np.pi*a_z**2)**(-1/4)*np.exp(-z**2/(2*a_z**2))

def prodVPsi0(z, rho, m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy):
    return V(rho, z, m, a_z, a_dd, r_vdw)*Psi0(z, a_z)**2

def intV(rho, m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy):
    return quad(prodVPsi0, a = -np.infty, b = np.infty, args = (rho, m, a_z, a_dd, r_vdw))
intV_all = np.vectorize(intV)

def findLocalMax(m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy, rho_min = 0.0001, rho_max = 10**2, n_points = 10**2):
    rho = np.linspace(rho_min, rho_max, n_points)
    integratedPotential = intV_all(rho, m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy)[0]
    result = argrelextrema(integratedPotential, np.greater) #np.less
    if result[0].size > 0: # not empty
        local_maxima = result[0][0]
        if (local_maxima > 0 and local_maxima < (n_points-1)):
            rho_min = rho[local_maxima - 1]
            rho_max = rho[local_maxima + 1]
            rho = np.linspace(rho_min, rho_max, n_points)
            integratedPotential = intV_all(rho, m, a_z, a_dd = a_dd_Dy, r_vdw = a_vdw_Dy)[0]
            local_maxima = argrelextrema(integratedPotential, np.greater)[0][0]
        else:
            print('Edge')
            return np.infty, 0
    else:
        print('No extrema found')
        return np.infty, 0
    return rho[local_maxima], integratedPotential[local_maxima]
```

```python
m = 0
A_z = np.linspace(1, 50, 100)
maxPotential = []
for a_z in A_z:
    result = findLocalMax(m, a_z)
    try:
        maxPotential += [result[1]]
    except:
        print(a_z, result)
        maxPotential += [0]
maxPotential = np.array(maxPotential)
```
