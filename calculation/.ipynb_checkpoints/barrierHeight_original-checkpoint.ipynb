```python
from scipy.optimize import root
import numpy as np
import matplotlib.pyplot as plt

def V(rho, z, alpha, m):
    return (m**2-1/4)/(2*rho**2)+3*(1-3*(z/rho)**2)/(z**2+rho**2)**(3/2)+1/2*alpha**4*z**2

def dVdz(rho, z, alpha, m):
    return z*(alpha**4+9*(z**2-3*rho**2)/(rho**2*(z**2+rho**2)**(5/2)))

def dVdrho(rho, z, alpha, m):
    return (1/4-m**2-9*rho**2*(-3*z**2+rho**2)/(z**2+rho**2)**(5/2)+18*z**2/(z**2+rho**2)**(3/2))/rho**3

def derivativeV(rho, z, alpha, m):
    return [dVdz(rho, z, alpha, m), dVdrho(rho, z, alpha, m)]
```

```python
def SaddlePosition(alpha, m):
    def currentDerivativeV(r):
        rho, z = r
        return derivativeV(rho, z, alpha, m)
    sol = root(currentDerivativeV, [1, 1])#, method='hybr')
    return sol.x
```

```python
plt.figure(dpi = 100)
plt.plot(AlphaList, SaddleHeight_M0, label = "m = 0")
plt.plot(AlphaList, SaddleHeight_M1, label = "m = 1")
plt.plot(AlphaList, SaddleHeight_M2, label = "m = 2")
plt.legend()
plt.xlabel("a_dd/a_z")
plt.ylabel("Barrier height (dipolar energy)")
plt.show()
```
